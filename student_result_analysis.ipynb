```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv("student.csv")
print(df.head())
```

```python
df.isnull().sum()
```

***Drop unnamed column***

```python
# Check if the column exists before attempting to drop it.
if "Unnamed: 0" in df.columns:
    df = df.drop("Unnamed: 0", axis=1)
    print(df.head())
else:
    print("Column 'Unnamed: 0' not found in DataFrame.")
print(df.head())
```

**Gender** **Distribution**

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

**From the above chart we have analyzed that no. of female is more than male**


```python
gb= df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot= True)
plt.show()
```

**From the above chart we have seen that parent education have a good impact on student result**

```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})

print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot= True)
plt.show()
```

**From the above chart we have seen parentmartialstatus don't have a huge impact (no impact/ negligible) on student's marks**

# **Distibution of Ethnic groups**

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
print(groupA)
```

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()
l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Ethnic Group Distribution")
plt.show()
```
